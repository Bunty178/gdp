```python
 

import pandas as pd
import numpy as ny
import chardet

def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc
 
 
my_encoding = find_encoding('NAD-Manipur-GSVA_cur_2014-15.csv')
df = pd.read_csv('NAD-Manipur-GSVA_cur_2014-15.csv', encoding=my_encoding)

find_encoding('NAD-Manipur-GSVA_cur_2014-15.csv')
```
